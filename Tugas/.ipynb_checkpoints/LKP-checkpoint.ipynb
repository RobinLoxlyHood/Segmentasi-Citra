```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread('ciplukan.jpg',0)
```

```python
scale_percent = 40 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
  
# resize image
img = cv.resize(img, dim, interpolation = cv.INTER_AREA)
```

# Deteksi Tepi dengan Sobel

```python
#Sobel X
sobelx64f = cv.Sobel(img,cv.CV_64F,1,0,ksize=3)
abs_sobelx64f = np.absolute(sobelx64f)
sobelx_8u = np.uint8(abs_sobelx64f)
#Sobel Y
sobely64f = cv.Sobel(img,cv.CV_64F,0,1,ksize=3)
abs_sobely64f = np.absolute(sobely64f)
sobely_8u = np.uint8(abs_sobely64f)
#Magnitude
magnitudesobel = cv.magnitude(sobelx64f,sobely64f)
abs_sobel64f = np.absolute(magnitudesobel)
sobel_8u = np.uint8(abs_sobel64f)
```

```python
cv.imshow('Sobel X', sobelx_8u)
cv.imshow('Sobel Y', sobely_8u)
cv.imshow('Sobel Magnitude', sobel_8u)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Deteksi Edge dengan Laplacian

```python
laplacian64f = cv.Laplacian(img,cv.CV_64F)
abs_laplacian64f = np.absolute(laplacian64f)
laplacian_8u = np.uint8(abs_laplacian64f)
```

```python
cv.imshow('laplacian', laplacian_8u)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Deteksi Edge dengan Canny Detection

```python
edges = cv.Canny(img,10,200)
```

```python
cv.imshow('Canny', edges)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Hasil

```python
plt.figure(figsize=(20, 30))
plt.subplot(3,3,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(3,3,2),plt.imshow(laplacian_8u,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(3,3,3),plt.imshow(sobelx_8u,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(3,3,4),plt.imshow(sobely_8u,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.subplot(3,3,5),plt.imshow(sobel_8u,cmap = 'gray')
plt.title('Sobel Magnitude'), plt.xticks([]), plt.yticks([])
plt.subplot(3,3,6),plt.imshow(edges,cmap = 'gray')
plt.title('Canny Detection'), plt.xticks([]), plt.yticks([])
plt.show()
```
